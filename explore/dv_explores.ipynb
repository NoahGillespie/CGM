## Patient 2

```python
pat_num = 2
print('High blood sugar threshold', np.percentile(data[pat_num].glu, 80))
df2 = max_glucose_dataset(cgm_data=data,participant_num=pat_num,glu_thresh=np.percentile(data[pat_num].glu, 80))
df2 = df2.dropna(subset=['diff_glucose'])
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty=None)
X = df2[['total_carb']]
y = df2['high_glucose']
## fit the model
log_reg.fit(X,y)
```

```python
plt.figure(figsize=(6,5))

plt.scatter(X, y)

plt.plot(np.linspace(0,160), log_reg.predict_proba(np.linspace(0,160).reshape(-1,1))[:,1], color = 'orange')
#plt.plot(np.linspace(7,13), my_log_reg.predict_proba(np.linspace(7,13).reshape(-1,1)), c = 'k')

plt.xlabel("total_carb", fontsize=12)
plt.ylabel("high_glucose", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df2['total_carb'],df2['diff_glucose'])

plt.show()
```

```python
slr = LinearRegression()
slr.fit(df2['total_carb'].values.reshape(-1,1),df2['diff_glucose'])
print(slr.intercept_)
print(slr.coef_)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df2['total_carb'],df2['diff_glucose'])
x = np.linspace(0,160)
plt.plot(x,slr.coef_*x + slr.intercept_,'r')

plt.show()
```

```python
# Linear regression on calories (eff. fat), total carb, protein
reg = LinearRegression()
reg.fit(df2[['calorie','total_carb','protein']].values, df2['diff_glucose'])
```

```python
reg.coef_, reg.intercept_
```
